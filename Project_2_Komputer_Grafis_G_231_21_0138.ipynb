```python
# Import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  Baca dataset komgraf.csv dengan pandas disini saya memakai variabel data
data = pd.read_csv('komgraf.csv')
```

```python
# membuat range
X = np.arange(1,13)

# membuat figur
fig = plt.figure(figsize=(10, 3))

# menentukan koordinat  untuk untuk bar chart
ax = fig.add_axes([0,0,1,1])

# membuat label
ax.set_title('Curah Hujan')
ax.set_xlabel('Bulan')
ax.set_ylabel('Curah Hujan (mm)')

# membuat bar chart
ax.bar(X + 0.00, data['1981-1990'], color = 'b', width = 0.25)
ax.bar(X + 0.25, data['1991-2000'], color = 'r', width = 0.25)
ax.bar(X + 0.50, data['2001-2010'], color = 'g', width = 0.25)

ax.legend(['1981-1990', '1991-2000', '2001-2010'], ncol=3, loc='upper center', frameon=False)

# Menambahkan garis horizontal
for d in [100, 200, 300, 400, 500, 600, 700]:
  if d != 700:
    plt.axhline(y=d, color='gray', linestyle='-', linewidth=0.3)
  else:
    plt.axhline(y=d, color='gray', linestyle=' ', linewidth=0.3)

# Mengatur skala sumbu x 1 - 12
ax.set_xticks(X + 0.25)
ax.set_xticklabels(np.arange(1, 13))

plt.show()
```

```python
#membaca folder csv
suhu = pd.read_csv("suhu_permukaan.csv")
#memberi nomer urut
suhu = suhu.set_index(pd.Index(range(1, len(suhu)+1)))
#menampilkan
suhu
```

```python
tipe = suhu['waktu']
asimilasi = suhu['asimilasi']
observasi = suhu['observasi']
non_asimilasi = suhu['non asimilasi']

# Plotting
plt.figure(figsize=(13, 5))
plt.plot(tipe, asimilasi, color='blue', label='Asimilasi',marker='*',zorder=2)
plt.plot(tipe, observasi, color='yellow', label='Observasi',marker='*',zorder=2)
plt.plot(tipe, non_asimilasi, color='red', label='Non Asimilasi',marker='*',zorder=2)

for temp in range(20, 28):
    plt.axhline(y=temp, color='gray', linestyle='-', linewidth=0.5,zorder=1)

# Kustomisasi
plt.xlabel('Waktu')
plt.ylabel('Suhu (°C)')
plt.title('Grafik Suhu')
plt.xticks(rotation=15)
plt.ylim(20, 27)

plt.legend()


# Tampilkan plot
plt.show()
```
